```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_diamonds = sns.load_dataset('diamonds')
df_diamonds_orig = df_diamonds.copy()
df_diamonds_orig.head()
```

```python
# prompt: OneHotEncoding для категориальных переменных. drop_first=True. Преобразовать True в 1 и Falce 0.

df_diamonds_orig = pd.get_dummies(df_diamonds_orig, columns=['cut', 'color', 'clarity'], drop_first=True)
df_diamonds_orig = df_diamonds_orig.replace({True: 1, False: 0})
df_diamonds_orig.head()
```

```python
df_diamonds_orig['carat_sq'] = df_diamonds_orig['carat']**2
df_diamonds_orig['carat_kub'] = df_diamonds_orig['carat']**3
df_diamonds_orig['depth_sq'] = df_diamonds_orig['depth']**2
df_diamonds_orig['table_sq'] = df_diamonds_orig['table']**2
df_diamonds_orig['x_sq'] = df_diamonds_orig['x']**2
df_diamonds_orig['x_kub'] = df_diamonds_orig['x']**3
df_diamonds_orig['y_sq'] = df_diamonds_orig['y']**2
df_diamonds_orig['y_kub'] = df_diamonds_orig['y']**3

# df_diamonds_orig['cut_rating_sq']=df_diamonds_orig['cut_rating']**2
# df_diamonds_orig['color_rating_sq']=df_diamonds_orig['color_rating']**2
# df_diamonds_orig['clarity_rating_sq']=df_diamonds_orig['clarity_rating']**2
```

```python
# X = df_diamonds_orig.drop(columns=['price', 'cut', 'color','clarity'])  # these are our "features" that we use to predict from
X = df_diamonds_orig.drop(columns=['price'])  # these are our "features" that we use to predict from
y = df_diamonds_orig['price']  # this is what we want to learn to predict
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
# # prompt: нормализовать значения переменных

# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_test_scaled = scaler.transform(X_test)
```

```python
#Split into training and testing sets

# Create and train the model
model_l = LinearRegression()
model_l.fit(X_train, y_train)

# Prediction
y_test_pred = model_l.predict(X_test)

# Evaluate the model
print(f'Model R^2 score: {model_l.score(X_test, y_test)}')

def RMSE(y_pred, y_true):
    return ((y_pred - y_true) ** 2).mean() ** 0.5

def summary(model_l):
    y_train_pred = model_l.predict(X_train)
    y_test_pred = model_l.predict(X_test)

    print("Train RMSE:", round(RMSE(y_train_pred, y_train),3))
    print("Test RMSE:", round(RMSE(y_test_pred, y_test),3))
    print("Train STD", round(y_train.std(),3))
    print("Test STD", round(y_test.std(), 3))
    return y_train_pred, y_test_pred


y_train_pred, y_test_pred = summary(model_l)
```

```python
# prompt: коэффициенты линейной модели в порядке убывания

coefs = pd.DataFrame({
    'feature': X.columns,
    'coef': model_l.coef_
}).sort_values(by='coef', ascending=False)

print(coefs)
```
